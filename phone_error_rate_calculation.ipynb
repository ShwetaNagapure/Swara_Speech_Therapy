To calculate the phone error rate (PER), you can use the same approach as in the Word Error Rate (WER) calculation but applied to phonemes. Phonemes are the basic units of sound in a language. Here's the modified code to include the calculation of the Phone Error Rate:

```python
from jiwer import wer
import speech_recognition as sr

def transcribe_audio(audio_path):
    recognizer = sr.Recognizer()

    with sr.AudioFile(audio_path) as source:
        audio_data = recognizer.record(source)

    # Replace this line with the appropriate speech recognition function
    transcribed_text = recognizer.recognize_google(audio_data, language='hi-IN')

    return transcribed_text

def calculate_per(reference_text, transcribed_text):
    # Tokenize phonemes in reference and transcribed text
    reference_phonemes = reference_text.split()
    transcribed_phonemes = transcribed_text.split()

    # Calculate PER
    substitutions = sum(1 for h, t in zip(reference_phonemes, transcribed_phonemes) if h != t)
    insertions = len(transcribed_phonemes) - len(reference_phonemes)
    deletions = len(reference_phonemes) - len(transcribed_phonemes)
    total_phonemes = len(reference_phonemes)

    per = (substitutions + insertions + deletions) / total_phonemes

    return per

def evaluate_transcription(reference_text, transcribed_text):
    # Calculate Word Error Rate (WER)
    wer_value = wer(reference_text, transcribed_text)

    # Calculate Phone Error Rate (PER)
    per = calculate_per(reference_text, transcribed_text)

    # You can add calculations for other metrics here

    return wer_value, per

def main():
    # Paths to your audio file and reference text
    audio_path = "/content/hi2501M0021.wav"
    reference_text = "धोबिना जाबा सोकारा उठाती तो देखति कि चौक साफ पड़ा हे और बरताना ममकजे हुए हैं"

    # Transcribe audio
    transcribed_text = transcribe_audio(audio_path)

    # Print the reference and transcribed text
    print("Reference Text:", reference_text)
    print("Transcribed Text:", transcribed_text)

    # Evaluate transcription
    wer_value, per = evaluate_transcription(reference_text, transcribed_text)

    # Print the results
    print(f"Word Error Rate: {wer_value*100 :.2f}%")
    print(f"Phone Error Rate: {per*100 :.2f}%")
    print(per)

if __name__ == "__main__":
    main()
```
